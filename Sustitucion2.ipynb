```python
def ngramas(texto, n):
    """Esta funcion divide el texto en ngramas"""
    texto = texto.replace(" ","").replace("\n","")
    return [texto[i:i+n] for i in range (len (texto) -n+1)]
```

```python
from collections import Counter
import pandas as pd
```

```python
def frec_ngramas (texto, n):
    """Calcula las frecuencias de los ngramas de un texto"""
    ng = ngramas(texto,n)
    frecuencias = Counter(ng)
    return pd.DataFrame(frecuencias.most_common(), columns = ['letra', 'frecuencia'])
```

```python
def leer_frecuencias(archivo):
    """Leer fichero con las frecuencias de los ngramas para el idioma español"""
    return pd.read_csv(archivo, sep = " ", names=['letra', 'frecuencia'])
```

```python
leer_frecuencias('spanish_monograms.txt')
```

```python
def leer_texto_cifrado(texto_cifrado, n):
    """Lee el texto cifrado y calcula las frecuencias de los ngramas"""
    with open (texto_cifrado, encoding='utf-8' ) as f:
        texto_cifrado = f.readlines()[0]
        return frec_ngramas(texto_cifrado, n)
```

## ALGORITMO DE DESCIFRADO

```python
def algoritmo_descifrado(texto_cifrado, alfabeto_permutado):
    """Descifra el texto cifrado a partir de un alfabeto permutado"""
    texto_plano = ""
    
    with open(texto_cifrado, encoding='utf-8') as f:
        texto_cifrado = f.read()

    for letra in texto_cifrado:
        if letra.lower() in alfabeto_permutado and alfabeto_permutado[letra.lower()] is not None:
            reemplazo = alfabeto_permutado[letra.lower()]
            if letra.isupper():
                texto_plano += reemplazo.upper()
            else:
                texto_plano += reemplazo
        else:
            texto_plano += letra

    return texto_plano
```

```python
algoritmo_descifrado("Texto_cifrado_2.txt", alfabeto_permutado)
```
